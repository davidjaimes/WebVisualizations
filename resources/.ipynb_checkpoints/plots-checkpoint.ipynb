```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
df = pd.read_csv('cities.csv')
df.head()
```

```python
plt.plot(df['Lat'], df['Max Temp'], 'o')
plt.xticks(np.arange(-60, 100, 20))
plt.yticks(np.arange(30, 135, 15))
plt.gca().xaxis.set_minor_locator(MultipleLocator(5))
plt.gca().yaxis.set_minor_locator(MultipleLocator(5))
plt.gca().tick_params(which='major', length=7, top=True, right=True)
plt.gca().tick_params(which='minor', length=4, top=True, right=True)
plt.xlim(-60, 80)
plt.ylim(30, 120)
plt.grid(linestyle='--')
plt.xlabel('Latitude (degrees)', size=16)
plt.ylabel(r'Max Temperature ($^{\circ}$F)', size=16)
plt.tight_layout()
plt.savefig('../images/temperature.png', dpi=200)
```

```python
plt.plot(df['Lat'], df['Humidity'], 'o')
plt.xticks(np.arange(-60, 100, 20))
plt.yticks(np.arange(0, 120, 15))
plt.gca().xaxis.set_minor_locator(MultipleLocator(5))
plt.gca().yaxis.set_minor_locator(MultipleLocator(5))
plt.gca().tick_params(which='major', length=7, top=True, right=True)
plt.gca().tick_params(which='minor', length=4, top=True, right=True)
plt.xlim(-60, 80)
plt.ylim(0, 105)
plt.grid(linestyle='--')
plt.xlabel('Latitude (degrees)', size=16)
plt.ylabel('Humidity (%)', size=16)
plt.tight_layout()
plt.savefig('../images/humidity.png', dpi=200)
```

```python
plt.plot(df['Lat'], df['Cloudiness'], 'o')
plt.xticks(np.arange(-60, 100, 20))
plt.yticks(np.arange(-10, 120, 15))
plt.gca().xaxis.set_minor_locator(MultipleLocator(5))
plt.gca().yaxis.set_minor_locator(MultipleLocator(5))
plt.gca().tick_params(which='major', length=7, top=True, right=True)
plt.gca().tick_params(which='minor', length=4, top=True, right=True)
plt.xlim(-60, 80)
plt.ylim(-10, 110)
plt.grid(linestyle='--')
plt.xlabel('Latitude (degrees)', size=16)
plt.ylabel('Cloudiness (%)', size=16)
plt.tight_layout()
plt.savefig('../images/cloudiness.png', dpi=200)
```

```python
plt.plot(df['Lat'], df['Wind Speed'], 'o')
plt.xticks(np.arange(-60, 100, 20))
plt.yticks(np.arange(0, 50, 10))
plt.gca().xaxis.set_minor_locator(MultipleLocator(5))
plt.gca().yaxis.set_minor_locator(MultipleLocator(2))
plt.gca().tick_params(which='major', length=7, top=True, right=True)
plt.gca().tick_params(which='minor', length=4, top=True, right=True)
plt.xlim(-60, 80)
plt.ylim(0, 40)
plt.grid(linestyle='--')
plt.xlabel('Latitude (degrees)', size=16)
plt.ylabel('Wind Speed (mph)', size=16)
plt.tight_layout()
plt.savefig('../images/wind-speed.png', dpi=200)
```
